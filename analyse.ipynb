```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the data from the text file into a DataFrame
file_path = 'results_LE_MANS_MORE.txt'  # Replace with the actual file path
df = pd.read_csv(file_path, sep=',', header=None, engine='python')

# Plot the data using histograms
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 8))

# Histogram for the first value
ax1.hist(df[0], bins=10, color='blue', edgecolor='black')
ax1.set_title('Distribution de la duree totale des simulations')
ax1.set_xlabel('Duree en ticks')
ax1.set_ylabel('Nombre de simulations')

# Calculate and display mean and standard deviation for the first value
mean_value_1 = df[0].mean()
std_value_1 = df[0].std()
ax1.axvline(mean_value_1, color='red', linestyle='dashed', linewidth=2, label=f'Moyenne: {mean_value_1:.2f}')
ax1.axvline(mean_value_1-std_value_1, color='green', linestyle='dashed', linewidth=2, label=f'Ecart Type: {std_value_1:.2f}')
ax1.axvline(mean_value_1+std_value_1, color='green', linestyle='dashed', linewidth=2)
ax1.legend()

# Histogram for the second value
ax2.hist(df[1], bins=10, color='orange', edgecolor='black')
ax2.set_title("Distribution de la duree moyenne d'un trajet")
ax2.set_xlabel('Duree en ticks')
ax2.set_ylabel('Nombre de simulations')

# Calculate and display mean and standard deviation for the second value
mean_value_2 = df[1].mean()
std_value_2 = df[1].std()
ax2.axvline(mean_value_2, color='red', linestyle='dashed', linewidth=2, label=f'Moyenne: {mean_value_2:.2f}')
ax2.axvline(mean_value_2-std_value_2, color='green', linestyle='dashed', linewidth=2, label=f'Ecart Type: {std_value_2:.2f}')
ax2.axvline(mean_value_2+std_value_2, color='green', linestyle='dashed', linewidth=2)
ax2.legend()

plt.tight_layout()
plt.show()
```
