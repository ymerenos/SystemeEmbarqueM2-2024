```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the data from the text file into a DataFrame
file_path = 'results_LE_MANS_FULL.txt'  # Replace with the actual file path
df = pd.read_csv(file_path, sep=',', header=None, engine='python')

# Plot the data using histograms
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 8))

# Histogram for the first value
ax1.hist(df[0], bins=10, color='blue', edgecolor='black')
ax1.set_title('Distribution de la duree totale des simulations')
ax1.set_xlabel('Duree en ticks')
ax1.set_ylabel('Nombre de simulations')

# Histogram for the second value
ax2.hist(df[1], bins=10, color='orange', edgecolor='black')
ax2.set_title('Distribution de la duree moyenne d\'un trajet')
ax2.set_xlabel('Duree en ticks')
ax2.set_ylabel('Nombre de simulations')

plt.tight_layout()
plt.show()
```
